# Data Analysis

```python
import os
import numpy as np
```

```python
# load all image data
data_path = r"D:\Repos\clip-bold\data\intermediate\02_stimulus_step"

images = []
for root, _, files in os.walk(data_path):
    for filename in files:
        print(filename)
        image = np.load(os.path.join(root, filename))
        images.append(np.load(os.path.join(root, filename)).astype(np.float16).flatten())
```

```python
images = np.vstack(images)
```

```python
images.shape
```

```python
images = images.astype(np.float32)
```

```python
# compute decomposition
from sklearn.decomposition import PCA

pca = PCA(40)
pca.fit(images)
```

```python
import matplotlib.pyplot as plt

# get explained variance
plt.plot(pca.explained_variance_ratio_)
plt.xlabel("Number of PCA Components")
plt.ylabel("Captured Variance Ratio")
plt.show()
```
